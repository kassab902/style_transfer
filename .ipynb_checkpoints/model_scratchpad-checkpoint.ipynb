```python
import tensorflow as tf
import tensorflow_addons as tfa
import tensorflow_datasets as tfds
import numpy as np
import pandas as pd
import sys,os,random,functools,math
```

```python
def load_data(batch_size = 16, image_shape = [256, 256]):
    def process(data):
        def _map(x, y):
            x = tf.image.resize(x,image_shape,method='nearest')
            x = tf.image.convert_image_dtype(x, tf.float32)
            return x,y
        data = data.map(_map)
        data = data.shuffle(1000)
        data = data.batch(batch_size)
        return data
    data = tfds.load('imagenette/320px-v2',as_supervised=True)
    train = data['train']
    validation = data['validation']

    return train.apply(process), validation.apply(process)
```

```python
train, valid = load_data()
train
```

```python
import matplotlib.pyplot as plt
def see_batch(batch,supervised=True,label_names=None):
    if supervised:
        batch, labels = batch
        labels = list(labels.numpy())
    n = len(batch)
    gridsize = int(math.sqrt(n))
    images = list(batch)
    fig, axes = plt.subplots(gridsize, gridsize)
    
    for image,axis,label in zip(images,np.ravel(axes),labels):
        axis.imshow(image)
        title = str(label_names[label] if label_names else label)
        axis.set_title(title)
        axis.set_axis_off()
```

```python
for batch in train:
    see_batch(batch)
#     print(x)
    break
```
